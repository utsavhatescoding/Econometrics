````python
print("Lets calculate the linear regression coeffecients i.e. the value of a and b  in the equation Y=a+bX")

DependantY=[]
IndependantX=[]
ob=int(input("Enter number of observations : "))
for i in range(1,ob+1):
  i1=input("Enter the explanatory variable X serially : ")
  i2=input("Enter the dependant variable Y serially : ")
  IndependantX.append(i1)
  DependantY.append(i2)

print(DependantY)
print(IndependantX)

````

````python
#Initialization :

sum_x=0
sum_y=0
sum_xsq=0
sum_xy=0


for i in range(0,ob):
  sum_x=sum_x+int(IndependantX[i])
  sum_y=sum_y+int(DependantY[i])
  sum_xsq=sum_xsq+(int(IndependantX[i]))**2
  sum_xy=sum_xy+(int(IndependantX[i])*int(DependantY[i]))

#Regression Coefficient

#a = ((sum_y * sum_xsq)-(sum_y*sum_xy))/((ob*sum_xsq)-(sum_x**2))

b= ((ob*sum_xy)-(sum_x*sum_y))/((ob*sum_xsq)-(sum_x**2))

a= (sum_y - b*sum_x)/ob

print("The intercept is a : ",a)
print("The slope is b : ",b)

print("The Regression Equation is Y = " + str(a) +"+" + str(b) + "X")

````

````python
#We can make predictions

X=float(input("Enter the value of explanatory variable X "))
Y=a+b*X
print("The predicted value of Y is : ",Y)
````
